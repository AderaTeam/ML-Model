```python
import pandas as pd
import seaborn as sb
import sktime
import matplotlib.pyplot as plt
import sweetviz as sv
from sktime.utils.plotting import plot_series
```

```python
GDP = pd.read_excel(
    'DATA/GDP.xlsx',
    parse_dates=['Дата '],
)
```

```python
GDP.set_index('Дата ')
```

```python
GDP['в млрд. рублей'] = GDP['в млрд. рублей'].apply(lambda a: float(str(a).replace(' ', '').replace(',', '.').replace('\xa0', '')))

```

```python
dataplot = sb.heatmap(
    GDP.corr(), 
    annot=True
)
```

```python
plt.plot(GDP['Дата '], GDP['в млрд. долл. США '] * 70, label='в млрд. долл. США ')
plt.plot(GDP['Дата '], GDP['в млрд. рублей'], label='в млрд. рублей')
plt.plot(GDP['Дата '], GDP['в процентах к ВВП**  '] * 100000, label='в млрд. рублей')
plt.legend()
```
